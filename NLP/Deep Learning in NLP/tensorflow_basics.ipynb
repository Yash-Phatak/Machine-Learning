```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
```

```python
# Creating a Linear Regression DataSet
N = 100 
X = np.random.random(N) *6 -3
y = 0.5*X-1+np.random.randn(N)*0.5
```

```python
plt.scatter(X,y) #Creates a Scatter Plot
```

```python
# Build Model 
i = Input(shape=(1,1)) # Single input feature
x = Dense(1)(i) # Dense Layer with 1 neuron. 
model =Model(i,x) # Output layer x is connected to Input Layer i
```

```python
model.summary()
```

```python
model.compile(
    loss = 'mse',
    optimizer = Adam(learning_rate=0.1),
    metrics=['mae']
)
```

```python
r = model.fit(
    X.reshape(-1,1),y,
    epochs = 200,
    batch_size = 32,
)
```

```python
# Plot loss per iteration
plt.plot(r.history['loss'],label='loss')
plt.legend();
```

```python
# Plot mae per iteration
plt.plot(r.history['mae'],label='mae')
plt.legend();
```

```python
# Make Predictions
Xtest = np.linspace(-3,3,20).reshape(-1,1)
ptest = model.predict(Xtest)
```

```python
plt.scatter(X,y)
plt.plot(Xtest,ptest)
```
